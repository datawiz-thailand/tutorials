Tutorial นี้จะแสดงตัวอย่างการใช้ Regression Tree และการทำ visualisation model

**Step 1) Load built-in data จาก sklearn;** tutorial นี้เลือกใช้ข้อมูล Boston house price  ซึ่งเป็นการทำ regression model

```python
from sklearn.datasets import load_boston

data = load_boston()
data.feature_names
```

```python
X = data.data
y = data.target
```

**Step 2) ทำ train test split**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

**Step 3) train Decision Tree และ predict ค่าในกลุ่ม test set**

```python
from sklearn import tree
regressor = tree.DecisionTreeRegressor(max_depth = 3, min_samples_leaf = 5)
```

```python
regressor.fit(X_train,y_train)
y_pred = regressor.predict(X_test)
```

**ตัวอย่าง classification tree**

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier(criterion='entropy',max_depth = 3, min_samples_leaf = 5)
```

**Step 4) ทำ model evaluation สำหรับกลุ่ม test set**

sklearn.metrics รวมเอา metric สำหรับการวัดผล model ไว้หลายหลาย สามารถเข้าไปดู doc และเลือกดูได้ครับ

```python
from sklearn.metrics import r2_score

print('r2_score: ',r2_score(y_test,y_pred))
```

**Step 5) export .dot file สำหรับทำ visualisation**

```python
feat_name = data.feature_names

dotfile = open('dtree.dot','w')

tree.export_graphviz(regressor,out_file=dotfile, feature_names=feat_name)

dotfile.close()
```
